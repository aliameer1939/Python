```python
import math                             
from scipy.spatial.distance import euclidean
from scipy.spatial.distance import chebyshev
from scipy.spatial import minkowski_distance
from scipy.spatial.distance import hamming
from scipy.spatial.distance import cosine
from scipy.stats import spearmanr
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style


class Similarityy:
#Similarity Method 1.................................................................................
    def  squared_euclidean_similarity(self,preference, id1, id2):
                                                       
        sum = 0
        for book in preference[id1]:                #for loop to iterate through each book
            if book in preference[id2]:             
                sum = euclidean(preference[id1][book][1], preference[id2][book][1]) 
        return sum





    #Similarity Method 2...................................................................................   
    def  minkowski_distance_similarity(self,preference, id1, id2):
        sum = 0                                                                                                                         #set the value of power to 1
        for book in preference[id1]:
            if book in preference[id2]:
            
                sum = minkowski_distance(preference[id1][book][1], preference[id2][book][1])
        return sum                                                                      




    #Similarity Method 3...................................................................................

    def  spearman_correlation_similarity(self,preference, id1, id2):
    
        step = 0                                        #inializing variables
        x=0
        y=0
        from math import sqrt
        n=1                                              #inialize n for count of iteration
        for books in preference[id1]:                     #for loop with if statement to check 2 books rated by one user or one book rated by 2 users  
            if books in preference[id2]:
                step=step+pow(abs(preference[id1][books][1] - preference[id2][books][1]),2)      # calculating square of x-y
                n=n+1

        x=6*step
        y=n*((n*n)-1)
        spearman_correlation = 1-(x*y)   
        return spearman_correlation





    #Similarity Methos 4......................................................................................
    def chebyshev_similarity(self,prefs, user1, user2):
        total=0                                                           
        for book in prefs[user1]:                                         
            if book in prefs[user2]:                                       #for loop with if statement to check 2 books rated by one user or one book rated by 2 users  
              
                sum = chebyshev(prefs[user1][book][1], prefs[user2][book][1])
        return total




    #Similarity Method 5.........................................................................................
    def hammming_distance_similarity(self,prefs, user1, user2):
        sum=0                                                 
        for book in prefs[user1]:                     # #for loop with if statement to check 2 books rated by one user or one book rated by 2 users  
            if book in prefs[user2]:
                sum = hamming(prefs[user1][book][1], prefs[user2][book][1])     
        return sum





    #Similarity Method 6..........................................................................................
    def  Cosine_similarity(self,preference, id1, id2):
        sum=0                         
        for book in preference[id1]:                    #for loop with if statement to check 2 books rated by one user or one book rated by 2 users  
            if book in preference[id2]:
                sum = 1-cosine(preference[id1][book][1], preference[id2][book][1])
                     
        return sum







    #Similarity Method 7.........................................................................................

    def  Pearson_Correlation(self,prefs, user1, user2):
        total=0                      #formula  (x2-x2)(y2-y1) divide by square roof of ((square of x2 -x1)(square of y2-y1))
        r=0
        xy=0                        #inializing variables to 0
        x2=0
        y2=0
        x2y2=0
        for book in prefs[user1]:                 #for loop and if statement to check all books rated by same user
            if book in prefs[user2]:
                xy= (prefs[user1][book][1])*(prefs[user2][book][1])
                x2=x2+pow(prefs[user1][book][1],2)                  #taking square using pow function
                y2=y2+pow(prefs[user2][book][1],2)
                x2y2=x2 * y2
        #print(x2y2)
        r = xy/1+(math.sqrt(x2y2))
        return r
    
    def plot(self,prefs,user1,user2):       
        style.use('grayscale')
        style.use('ggplot')
        a=(self.Pearson_Correlation(prefs, user1, user2),self.hammming_distance_similarity(prefs, user1, user2),self.Cosine_similarity(prefs, user1, user2),self.chebyshev_similarity(prefs, user1, user2),self.spearman_correlation_similarity(prefs, user1, user2),self.minkowski_distance_similarity(prefs, user1, user2),self.squared_euclidean_similarity(prefs, user1, user2))
        #a = [1,2,3,4,5,6,7,8,9]
        plt.plot(a)
        plt.title("Line Graph")
        plt.xlabel("Similarity Score")
        
```
